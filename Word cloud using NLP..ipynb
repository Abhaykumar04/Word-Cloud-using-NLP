### TOKENIZATION:

It is the process of breaking down a piece of text into individual words or "tokens". It is a common pre-processing step in natural language processing (NLP) tasks, such as text classification and sentiment analysis. Tokenization is usually performed by splitting the text on whitespace and punctuation characters, although more advanced methods that take into account the context and structure of the text are also available.

```python
tokenized_word = word_tokenize (paragraph)
tokenized_word
```

```python
tokenized_sent = sent_tokenize (paragraph)
tokenized_sent
```

```python
nltk.download("stopwords")
```

```python
tokens_without_sw = [word for word in tokenized_word if not word in stopwords.words()]
sent_without_sw = [word for word in tokenized_sent if not word in stopwords.words()]
```

```python
print(tokens_without_sw)
```

```python
print(sent_without_sw)
```

```python
filtered_sentence = (" ").join(tokens_without_sw)
print(filtered_sentence)
```

```python
text = re.sub("[,. ?]", " ", filtered_sentence)
text
```

### Stemming:
It is the process of reducing words to their base or root form to simplify text analysis.

#### Under-stemming:  False negative
It is the failure to reduce related words to the same stem.
#### over-stemming:  False positive
It is the reduction of unrelated words to the same stem.

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

for i in tokenized_word:
    print(i,":",ps.stem(i))
```

### Lemmatization:
It is a text preprocessing technique that involves reducing words to their base form, or lemma, based on their intended meaning. It uses a dictionary or knowledge base to convert words to their base form, which can improve text analysis accuracy by reducing the number of variations of a word. Unlike stemming, lemmatization produces valid words that can be found in a dictionary.

```python
lemmatizer = WordNetLemmatizer()
lemmatized_output = ' '.join([lemmatizer.lemmatize(w) for w in tokenized_word])
print(lemmatized_output)
```

####  Word cloud

It is a visual representation of text data in which the most frequent words are displayed in larger fonts, while less frequent words are displayed in smaller fonts. It is a popular tool for quickly identifying and visualizing the most prominent words in a given text dataset.

```python
comment_words = ''
stopwords = set(STOPWORDS)
 
# Loop through each tokenized word
for val in tokenized_word:
    
    # typecaste each val to string
    val = str(val)
    tokens = val.split()
     
    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
     
    comment_words += " ".join(tokens)+" "
 
wordcloud = WordCloud(width = 1600, height = 800,
                background_color ='black',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
 
# plot the WordCloud image                      
plt.figure(figsize = (14, 20), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
